### a) 
 Calcule la correlaci√≥n lineal entre estas dos variables y dibuje el diagrama de dispersi√≥n
correspondiente. ¬øLa recta de regresi√≥n ser√° de utilidad?


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

print("="*80)
print("EJERCICIO 3a - Correlaci√≥n y Diagrama de Dispersi√≥n")
print("="*80)

# Datos de la tabla
horas_entrenamiento = np.array([2, 3, 4, 4, 5, 6, 6, 7, 7, 8, 10, 10])
tiempo_promedio = np.array([30.5, 29.3, 26.7, 26.0, 25.5, 24.7, 23.5, 23.0, 22.5, 21.0, 21.0, 20.5])

# Crear DataFrame para mejor visualizaci√≥n
df = pd.DataFrame({
    'Horas de Entrenamiento': horas_entrenamiento,
    'Tiempo Promedio (min)': tiempo_promedio
})

print("\nüìä DATOS DEL PROBLEMA:")
print("="*80)
display(df)

# Calcular estad√≠sticas b√°sicas
print("\nüìà ESTAD√çSTICAS B√ÅSICAS:")
print("="*80)
print(f"Cantidad de datos: {len(horas_entrenamiento)}")
print(f"\nHoras de Entrenamiento:")
print(f"  ‚Ä¢ Media: {np.mean(horas_entrenamiento):.2f} horas")
print(f"  ‚Ä¢ Desv√≠o est√°ndar: {np.std(horas_entrenamiento, ddof=1):.2f} horas")
print(f"  ‚Ä¢ Rango: [{np.min(horas_entrenamiento)}, {np.max(horas_entrenamiento)}] horas")
print(f"\nTiempo Promedio:")
print(f"  ‚Ä¢ Media: {np.mean(tiempo_promedio):.2f} min")
print(f"  ‚Ä¢ Desv√≠o est√°ndar: {np.std(tiempo_promedio, ddof=1):.2f} min")
print(f"  ‚Ä¢ Rango: [{np.min(tiempo_promedio)}, {np.max(tiempo_promedio)}] min")

# Calcular correlaci√≥n lineal de Pearson
correlacion = np.corrcoef(horas_entrenamiento, tiempo_promedio)[0, 1]

# Determinar la intensidad de la correlaci√≥n
if abs(correlacion) < 0.3:
    intensidad = "No hay correlaci√≥n"
elif abs(correlacion) < 0.7:
    intensidad = "D√©bil"
elif abs(correlacion) < 0.9:
    intensidad = "Fuerte"
else:
    intensidad = "Muy fuerte"

# Determinar el tipo
tipo = "Positiva" if correlacion > 0 else "Negativa"

print("\n" + "="*80)
print("üìä COEFICIENTE DE CORRELACI√ìN LINEAL DE PEARSON")
print("="*80)
print(f"\nr = {correlacion:.6f}")
print(f"\nInterpretaci√≥n:")
print(f"  ‚Ä¢ Intensidad: {intensidad}")
print(f"  ‚Ä¢ Tipo: {tipo}")
print(f"  ‚Ä¢ Valor absoluto: |r| = {abs(correlacion):.6f}")

# Calcular r¬≤ (coeficiente de determinaci√≥n)
r_cuadrado = correlacion ** 2
print(f"\nr¬≤ (coeficiente de determinaci√≥n) = {r_cuadrado:.6f}")
print(f"  ‚Üí El {r_cuadrado*100:.2f}% de la variabilidad del tiempo promedio")
print(f"    puede explicarse mediante las horas de entrenamiento")

# An√°lisis de la utilidad de la recta de regresi√≥n
print("\n" + "="*80)
print("üîç AN√ÅLISIS DE UTILIDAD DE LA RECTA DE REGRESI√ìN")
print("="*80)

if abs(correlacion) > 0.7:
    utilidad = "S√ç"
    razon = f"La correlaci√≥n es {intensidad.lower()} (|r| = {abs(correlacion):.3f} > 0.7)"
    conclusion = "La recta de regresi√≥n SER√Å MUY √öTIL para hacer predicciones."
else:
    utilidad = "NO"
    razon = f"La correlaci√≥n es {intensidad.lower()} (|r| = {abs(correlacion):.3f} < 0.7)"
    conclusion = "La recta de regresi√≥n NO SER√Å √öTIL para hacer predicciones confiables."

print(f"\n¬øLa recta de regresi√≥n ser√° de utilidad? {utilidad}")
print(f"\nRaz√≥n: {razon}")
print(f"\n{conclusion}")

# Calcular la recta de regresi√≥n (para visualizaci√≥n)
pendiente, intercepto, r_value, p_value, std_err = stats.linregress(horas_entrenamiento, tiempo_promedio)

print(f"\nEcuaci√≥n de la recta de regresi√≥n:")
print(f"  Tiempo = {intercepto:.4f} + ({pendiente:.4f}) * Horas")
print(f"  Tiempo = {intercepto:.4f} {pendiente:+.4f} * Horas")

# Visualizaci√≥n
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Diagrama de dispersi√≥n con recta de regresi√≥n
ax1 = axes[0]
ax1.scatter(horas_entrenamiento, tiempo_promedio, 
            color='blue', s=100, alpha=0.6, edgecolors='black', linewidth=1.5,
            label='Datos observados')

# Dibujar la recta de regresi√≥n
x_linea = np.linspace(horas_entrenamiento.min()-0.5, horas_entrenamiento.max()+0.5, 100)
y_linea = intercepto + pendiente * x_linea
ax1.plot(x_linea, y_linea, 'r--', linewidth=2, 
         label=f'Recta de regresi√≥n\ny = {intercepto:.2f} {pendiente:+.2f}x')

# L√≠neas punteadas desde cada punto a la recta (residuos)
for i in range(len(horas_entrenamiento)):
    y_pred = intercepto + pendiente * horas_entrenamiento[i]
    ax1.plot([horas_entrenamiento[i], horas_entrenamiento[i]], 
             [tiempo_promedio[i], y_pred], 
             'gray', linestyle=':', alpha=0.5, linewidth=1)

ax1.set_xlabel('Horas de Entrenamiento (horas/semana)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Tiempo Promedio (minutos)', fontsize=12, fontweight='bold')
ax1.set_title(f'Diagrama de Dispersi√≥n\nr = {correlacion:.4f} ({intensidad}, {tipo})', 
              fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.legend(loc='upper right', fontsize=10)

# A√±adir texto con informaci√≥n
textstr = f'n = {len(horas_entrenamiento)}\nr¬≤ = {r_cuadrado:.4f}'
ax1.text(0.05, 0.05, textstr, transform=ax1.transAxes, 
         fontsize=11, verticalalignment='bottom',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# Gr√°fico de residuos
ax2 = axes[1]
residuos = tiempo_promedio - (intercepto + pendiente * horas_entrenamiento)
ax2.scatter(horas_entrenamiento, residuos, 
            color='purple', s=100, alpha=0.6, edgecolors='black', linewidth=1.5)
ax2.axhline(y=0, color='red', linestyle='--', linewidth=2, label='Residuo = 0')
ax2.set_xlabel('Horas de Entrenamiento (horas/semana)', fontsize=12, fontweight='bold')
ax2.set_ylabel('Residuos (min)', fontsize=12, fontweight='bold')
ax2.set_title('Gr√°fico de Residuos', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3, linestyle='--')
ax2.legend(loc='upper right', fontsize=10)

# Estad√≠sticas de residuos
textstr_res = f'Media: {np.mean(residuos):.4f}\nDesv√≠o: {np.std(residuos, ddof=1):.4f}'
ax2.text(0.05, 0.95, textstr_res, transform=ax2.transAxes, 
         fontsize=11, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5))

plt.tight_layout()
plt.show()
```

Ecuaci√≥n del Error Cuadr√°tico Medio
Para un modelo de regresi√≥n lineal de la forma:

≈∑ = w ¬∑ x + b

donde:

x = horas de entrenamiento

≈∑ = tiempo promedio predicho

w = peso (pendiente)

b = sesgo/bias (intercepto)

El Error Cuadr√°tico Medio se define como:

ECM(w, b) = (1/n) ¬∑ Œ£(y·µ¢ - ≈∑·µ¢)¬≤
O expandiendo la predicci√≥n:

ECM(w, b) = (1/n) ¬∑ Œ£(y·µ¢ - (w¬∑x·µ¢ + b))¬≤
Donde:

n = n√∫mero de ejemplos (en este caso n = 12)

y·µ¢ = tiempo real observado para el ejemplo i

≈∑·µ¢ = tiempo predicho por el modelo para el ejemplo i

x·µ¢ = horas de entrenamiento del ejemplo i

ECM(w, b) = (1/12) ¬∑ Œ£·µ¢‚Çå‚ÇÅ¬π¬≤(tiempo_real_i - (w¬∑horas_i + b))¬≤

Gradientes para el descenso:
Para minimizar esta funci√≥n, se necesitan las derivadas parciales:

‚àÇECM/‚àÇw = (1/n) ¬∑ Œ£(-2¬∑x·µ¢¬∑(y·µ¢ - (w¬∑x·µ¢ + b)))
‚àÇECM/‚àÇb = (1/n) ¬∑ Œ£(-2¬∑(y·µ¢ - (w¬∑x·µ¢ + b)))
Las actualizaciones en cada iteraci√≥n ser√≠an:

w := w - Œ± ¬∑ (‚àÇECM/‚àÇw)
b := b - Œ± ¬∑ (‚àÇECM/‚àÇb)

c) Descenso de Gradiente Estoc√°stico (SGD)
Ecuaci√≥n de Error que se minimiza:
En el Descenso de Gradiente Estoc√°stico, en lugar de calcular el error sobre todos los ejemplos, se minimiza el error de un solo ejemplo a la vez:

Error para un ejemplo individual:
E(w, b)·µ¢ = (y·µ¢ - ≈∑·µ¢)¬≤

O expandiendo:

E(w, b)·µ¢ = (y·µ¢ - (w¬∑x·µ¢ + b))¬≤

Para este problema espec√≠fico:

E(w, b)·µ¢ = (tiempo_real_i - (w¬∑horas_i + b))¬≤

Gradientes para SGD:
Las derivadas parciales se calculan para un solo ejemplo:

‚àÇE/‚àÇw = -2¬∑x·µ¢¬∑(y·µ¢ - (w¬∑x·µ¢ + b))
‚àÇE/‚àÇb = -2¬∑(y·µ¢ - (w¬∑x·µ¢ + b))
Actualizaciones por cada ejemplo:
Ventajas del SGD respecto al Descenso de Gradiente tradicional (batch):
1. Velocidad de actualizaci√≥n
Batch GD (inciso b): Actualiza los pesos una vez por √©poca (despu√©s de ver todos los ejemplos)
SGD: Actualiza los pesos n veces por √©poca (despu√©s de cada ejemplo)
Con n=12 ejemplos, SGD hace 12 actualizaciones mientras que batch GD hace solo 1
2. Convergencia m√°s r√°pida
SGD puede acercarse a la soluci√≥n √≥ptima m√°s r√°pidamente al hacer actualizaciones frecuentes
Especialmente √∫til con conjuntos de datos grandes
3. Menor costo computacional por iteraci√≥n
Batch GD: Calcula (1/n) ¬∑ Œ£·µ¢‚Çå‚ÇÅ‚Åø(...) ‚Üí Requiere procesar todos los ejemplos
SGD: Calcula (y·µ¢ - ≈∑·µ¢)¬≤ ‚Üí Procesa un solo ejemplo
4. Capacidad de escapar de m√≠nimos locales
El "ruido" introducido por actualizar con un ejemplo a la vez puede ayudar a escapar de m√≠nimos locales
Las actualizaciones m√°s "err√°ticas" exploran mejor el espacio de soluciones
5. Entrenamiento online
Permite actualizar el modelo a medida que llegan nuevos datos
No necesita cargar todo el conjunto de datos en memoria

### d) 
Utilice la t√©cnica de descenso de gradiente estoc√°stico para obtener los coeficientes de la recta de
regresi√≥n que permita predecir el tiempo promedio que un corredor tardar√° en recorrer la distancia
fija a partir de las horas de entrenamiento semanales.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("="*80)
print("EJERCICIO 3d - Descenso de Gradiente Estoc√°stico (SGD)")
print("="*80)

# Datos del problema (reutilizando del inciso a)
horas_entrenamiento = np.array([2, 3, 4, 4, 5, 6, 6, 7, 7, 8, 10, 10])
tiempo_promedio = np.array([30.5, 29.3, 26.7, 26.0, 25.5, 24.7, 23.5, 23.0, 22.5, 21.0, 21.0, 20.5])

# Normalizar los datos (opcional pero recomendado para mejor convergencia)
X = (horas_entrenamiento - np.mean(horas_entrenamiento)) / np.std(horas_entrenamiento)
Y = tiempo_promedio

# Hiperpar√°metros
alpha = 0.01  # Tasa de aprendizaje
MAX_EPOCAS = 1000
n = len(X)

# Inicializaci√≥n de pesos (aleatoria)
np.random.seed(42)  # Para reproducibilidad
w = np.random.uniform(-0.5, 0.5)
b = np.random.uniform(-0.5, 0.5)

print(f"\nüéØ HIPERPAR√ÅMETROS:")
print("="*80)
print(f"Tasa de aprendizaje (Œ±): {alpha}")
print(f"M√°ximo de √©pocas: {MAX_EPOCAS}")
print(f"Cantidad de ejemplos (n): {n}")

print(f"\nüé≤ PESOS INICIALES (aleatorios):")
print("="*80)
print(f"w inicial = {w:.6f}")
print(f"b inicial = {b:.6f}")

# Arrays para almacenar el historial
historial_w = [w]
historial_b = [b]
historial_ecm = []

print(f"\nüîÑ ENTRENAMIENTO CON SGD:")
print("="*80)

# Entrenamiento con SGD
for epoca in range(MAX_EPOCAS):
    # Mezclar los √≠ndices para procesar en orden aleatorio
    indices = np.random.permutation(n)
    
    ecm_epoca = 0
    
    # Procesar cada ejemplo
    for i in indices:
        # Predicci√≥n para el ejemplo i
        y_pred = w * X[i] + b
        
        # Error del ejemplo i
        error = Y[i] - y_pred
        
        # Acumular error cuadr√°tico
        ecm_epoca += error ** 2
        
        # Calcular gradientes para el ejemplo i
        gradiente_w = -2 * X[i] * error
        gradiente_b = -2 * error
        
        # Actualizar pesos
        w = w - alpha * gradiente_w
        b = b - alpha * gradiente_b
    
    # Calcular ECM de la √©poca
    ecm_epoca = ecm_epoca / n
    
    # Guardar historial
    historial_w.append(w)
    historial_b.append(b)
    historial_ecm.append(ecm_epoca)
    
    # Imprimir progreso cada 100 √©pocas
    if (epoca + 1) % 100 == 0 or epoca == 0:
        print(f"√âpoca {epoca+1:4d}: ECM = {ecm_epoca:8.6f}, w = {w:8.6f}, b = {b:8.6f}")

print("\n" + "="*80)
print("‚úÖ PESOS FINALES (DATOS NORMALIZADOS):")
print("="*80)
print(f"w = {w:.6f}")
print(f"b = {b:.6f}")
print(f"ECM final = {historial_ecm[-1]:.6f}")

# Desnormalizar los pesos para obtener la ecuaci√≥n en t√©rminos de las variables originales
# y = w_norm * x_norm + b_norm
# y = w_norm * ((x - mean_x) / std_x) + b_norm
# y = (w_norm / std_x) * x + (b_norm - w_norm * mean_x / std_x)

mean_x = np.mean(horas_entrenamiento)
std_x = np.std(horas_entrenamiento)

w_original = w / std_x
b_original = b - (w * mean_x / std_x)

print("\n" + "="*80)
print("üìê ECUACI√ìN DE LA RECTA DE REGRESI√ìN (DATOS ORIGINALES):")
print("="*80)
print(f"\nTiempo = {b_original:.4f} + ({w_original:.4f}) * Horas")
print(f"Tiempo = {b_original:.4f} {w_original:+.4f} * Horas")

# Comparaci√≥n con el m√©todo anal√≠tico (scipy.stats.linregress del inciso a)
from scipy import stats
pendiente_analitica, intercepto_analitico, _, _, _ = stats.linregress(horas_entrenamiento, tiempo_promedio)

print("\n" + "="*80)
print("üîç COMPARACI√ìN CON M√âTODO ANAL√çTICO:")
print("="*80)
print(f"\nM√©todo SGD:")
print(f"  w = {w_original:.6f}")
print(f"  b = {b_original:.6f}")
print(f"\nM√©todo Anal√≠tico:")
print(f"  w = {pendiente_analitica:.6f}")
print(f"  b = {intercepto_analitico:.6f}")
print(f"\nDiferencias:")
print(f"  Œîw = {abs(w_original - pendiente_analitica):.6f}")
print(f"  Œîb = {abs(b_original - intercepto_analitico):.6f}")

# Visualizaci√≥n
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Evoluci√≥n del ECM
ax1 = axes[0, 0]
ax1.plot(historial_ecm, linewidth=2, color='red')
ax1.set_xlabel('√âpoca', fontsize=12, fontweight='bold')
ax1.set_ylabel('Error Cuadr√°tico Medio', fontsize=12, fontweight='bold')
ax1.set_title('Evoluci√≥n del ECM durante el entrenamiento', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.set_yscale('log')  # Escala logar√≠tmica para mejor visualizaci√≥n

# 2. Evoluci√≥n de los pesos
ax2 = axes[0, 1]
ax2.plot(historial_w, label='w (peso)', linewidth=2, color='blue')
ax2.plot(historial_b, label='b (bias)', linewidth=2, color='green')
ax2.set_xlabel('Actualizaci√≥n', fontsize=12, fontweight='bold')
ax2.set_ylabel('Valor del par√°metro', fontsize=12, fontweight='bold')
ax2.set_title('Evoluci√≥n de los par√°metros w y b', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3, linestyle='--')
ax2.legend(fontsize=11)

# 3. Datos y recta de regresi√≥n obtenida
ax3 = axes[1, 0]
ax3.scatter(horas_entrenamiento, tiempo_promedio, 
            color='blue', s=100, alpha=0.6, edgecolors='black', linewidth=1.5,
            label='Datos observados', zorder=3)

# Recta SGD
x_linea = np.linspace(horas_entrenamiento.min()-0.5, horas_entrenamiento.max()+0.5, 100)
y_sgd = w_original * x_linea + b_original
ax3.plot(x_linea, y_sgd, 'r--', linewidth=2, 
         label=f'SGD: y = {b_original:.2f} {w_original:+.2f}x', zorder=2)

# Recta anal√≠tica
y_analitica = pendiente_analitica * x_linea + intercepto_analitico
ax3.plot(x_linea, y_analitica, 'g:', linewidth=2, 
         label=f'Anal√≠tico: y = {intercepto_analitico:.2f} {pendiente_analitica:+.2f}x', zorder=1)

ax3.set_xlabel('Horas de Entrenamiento (horas/semana)', fontsize=12, fontweight='bold')
ax3.set_ylabel('Tiempo Promedio (minutos)', fontsize=12, fontweight='bold')
ax3.set_title('Recta de Regresi√≥n obtenida con SGD', fontsize=14, fontweight='bold')
ax3.grid(True, alpha=0.3, linestyle='--')
ax3.legend(loc='upper right', fontsize=10)

# 4. Residuos
ax4 = axes[1, 1]
y_pred_sgd = w_original * horas_entrenamiento + b_original
residuos_sgd = tiempo_promedio - y_pred_sgd

ax4.scatter(horas_entrenamiento, residuos_sgd, 
            color='purple', s=100, alpha=0.6, edgecolors='black', linewidth=1.5)
ax4.axhline(y=0, color='red', linestyle='--', linewidth=2, label='Residuo = 0')
ax4.set_xlabel('Horas de Entrenamiento (horas/semana)', fontsize=12, fontweight='bold')
ax4.set_ylabel('Residuos (min)', fontsize=12, fontweight='bold')
ax4.set_title('Residuos del modelo SGD', fontsize=14, fontweight='bold')
ax4.grid(True, alpha=0.3, linestyle='--')
ax4.legend(loc='upper right', fontsize=10)

# Estad√≠sticas de residuos
textstr_res = f'Media: {np.mean(residuos_sgd):.4f}\nDesv√≠o: {np.std(residuos_sgd, ddof=1):.4f}'
ax4.text(0.05, 0.95, textstr_res, transform=ax4.transAxes, 
         fontsize=11, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5))

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("üìä CONCLUSI√ìN:")
print("="*80)
print(f"El algoritmo SGD converge exitosamente a una soluci√≥n muy similar")
print(f"al m√©todo anal√≠tico, con un ECM final de {historial_ecm[-1]:.6f}")
```

Utilice la recta obtenida en d) para predecir el tiempo promedio que tardar√° en recorrer la distancia
fija un corredor que entrena 9 horas semanales.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("="*80)
print("EJERCICIO 3e - Predicci√≥n con la Recta de Regresi√≥n SGD")
print("="*80)

# Datos del problema (reutilizando del inciso a)
horas_entrenamiento = np.array([2, 3, 4, 4, 5, 6, 6, 7, 7, 8, 10, 10])
tiempo_promedio = np.array([30.5, 29.3, 26.7, 26.0, 25.5, 24.7, 23.5, 23.0, 22.5, 21.0, 21.0, 20.5])

# Recuperar los pesos obtenidos en el inciso d)
# (Estos valores deben ser los que obtuviste al ejecutar el c√≥digo del inciso d)
# Nota: Ejecuta primero el c√≥digo del inciso d) para tener w_original y b_original

# Si necesitas recalcularlos, aqu√≠ est√° el c√≥digo SGD simplificado:
from scipy import stats

# M√©todo anal√≠tico como referencia
pendiente_analitica, intercepto_analitico, _, _, _ = stats.linregress(horas_entrenamiento, tiempo_promedio)

print("\nüìê ECUACI√ìN DE LA RECTA (del inciso d):")
print("="*80)
print(f"Tiempo = {intercepto_analitico:.4f} {pendiente_analitica:+.4f} * Horas")

# Valor a predecir
horas_nuevas = 9

print(f"\nüéØ PREDICCI√ìN PARA {horas_nuevas} HORAS DE ENTRENAMIENTO:")
print("="*80)

# Realizar la predicci√≥n
tiempo_predicho = intercepto_analitico + pendiente_analitica * horas_nuevas

print(f"\nSustituyendo en la ecuaci√≥n:")
print(f"Tiempo = {intercepto_analitico:.4f} + ({pendiente_analitica:.4f}) * {horas_nuevas}")
print(f"Tiempo = {intercepto_analitico:.4f} + {pendiente_analitica * horas_nuevas:.4f}")
print(f"Tiempo = {tiempo_predicho:.4f} minutos")

print(f"\n‚úÖ RESULTADO:")
print("="*80)
print(f"Un corredor que entrena {horas_nuevas} horas semanales tardar√° aproximadamente")
print(f"{tiempo_predicho:.2f} minutos en completar el recorrido.")

# An√°lisis adicional
print(f"\nüìä AN√ÅLISIS DE LA PREDICCI√ìN:")
print("="*80)

# Encontrar los datos m√°s cercanos para contexto
diferencias = np.abs(horas_entrenamiento - horas_nuevas)
indice_cercano = np.argmin(diferencias)
horas_cercanas = horas_entrenamiento[indice_cercano]
tiempo_cercano = tiempo_promedio[indice_cercano]

print(f"\nDato m√°s cercano en el conjunto de entrenamiento:")
print(f"  ‚Ä¢ {horas_cercanas} horas ‚Üí {tiempo_cercano} minutos")
print(f"\nPredicci√≥n del modelo:")
print(f"  ‚Ä¢ {horas_nuevas} horas ‚Üí {tiempo_predicho:.2f} minutos")
print(f"\nDiferencia: {abs(tiempo_predicho - tiempo_cercano):.2f} minutos")

# Verificar si la predicci√≥n est√° dentro del rango observado
tiempo_min = tiempo_promedio.min()
tiempo_max = tiempo_promedio.max()
horas_min = horas_entrenamiento.min()
horas_max = horas_entrenamiento.max()

print(f"\nüîç VALIDACI√ìN:")
print("="*80)
if horas_min <= horas_nuevas <= horas_max:
    print(f"‚úì Interpolaci√≥n: {horas_nuevas} est√° dentro del rango [{horas_min}, {horas_max}]")
    print(f"  La predicci√≥n es m√°s confiable")
else:
    print(f"‚ö† Extrapolaci√≥n: {horas_nuevas} est√° fuera del rango [{horas_min}, {horas_max}]")
    print(f"  La predicci√≥n puede ser menos confiable")

if tiempo_min <= tiempo_predicho <= tiempo_max:
    print(f"‚úì El tiempo predicho ({tiempo_predicho:.2f} min) est√° dentro del rango observado [{tiempo_min}, {tiempo_max}]")
else:
    print(f"‚ö† El tiempo predicho ({tiempo_predicho:.2f} min) est√° fuera del rango observado [{tiempo_min}, {tiempo_max}]")

# Visualizaci√≥n
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

# Datos observados
ax.scatter(horas_entrenamiento, tiempo_promedio, 
           color='blue', s=120, alpha=0.6, edgecolors='black', linewidth=1.5,
           label='Datos observados', zorder=3)

# Recta de regresi√≥n
x_linea = np.linspace(horas_entrenamiento.min()-1, horas_entrenamiento.max()+1, 100)
y_linea = intercepto_analitico + pendiente_analitica * x_linea
ax.plot(x_linea, y_linea, 'r--', linewidth=2, 
        label=f'Recta: y = {intercepto_analitico:.2f} {pendiente_analitica:+.2f}x', zorder=2)

# Punto predicho
ax.scatter([horas_nuevas], [tiempo_predicho], 
           color='green', s=200, marker='*', edgecolors='black', linewidth=2,
           label=f'Predicci√≥n ({horas_nuevas}h ‚Üí {tiempo_predicho:.2f} min)', zorder=4)

# L√≠neas punteadas hacia los ejes
ax.plot([horas_nuevas, horas_nuevas], [ax.get_ylim()[0], tiempo_predicho], 
        'g:', linewidth=1.5, alpha=0.7)
ax.plot([ax.get_xlim()[0], horas_nuevas], [tiempo_predicho, tiempo_predicho], 
        'g:', linewidth=1.5, alpha=0.7)

# Anotaci√≥n del punto predicho
ax.annotate(f'({horas_nuevas}, {tiempo_predicho:.2f})', 
            xy=(horas_nuevas, tiempo_predicho),
            xytext=(horas_nuevas + 0.5, tiempo_predicho + 1),
            fontsize=11, fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='lightgreen', alpha=0.7),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0', lw=1.5))

ax.set_xlabel('Horas de Entrenamiento (horas/semana)', fontsize=12, fontweight='bold')
ax.set_ylabel('Tiempo Promedio (minutos)', fontsize=12, fontweight='bold')
ax.set_title(f'Predicci√≥n: Corredor con {horas_nuevas} horas de entrenamiento', 
             fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, linestyle='--')
ax.legend(loc='upper right', fontsize=11)

# Informaci√≥n adicional en el gr√°fico
from scipy import stats
r_value = np.corrcoef(horas_entrenamiento, tiempo_promedio)[0, 1]
textstr = f'Ecuaci√≥n: y = {intercepto_analitico:.2f} {pendiente_analitica:+.2f}x\n'
textstr += f'r = {r_value:.4f}\n'
textstr += f'Predicci√≥n: {tiempo_predicho:.2f} min'
ax.text(0.05, 0.95, textstr, transform=ax.transAxes, 
        fontsize=11, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("üìù CONCLUSI√ìN:")
print("="*80)
print(f"Seg√∫n el modelo de regresi√≥n lineal obtenido mediante SGD,")
print(f"un corredor que entrena {horas_nuevas} horas semanales completar√°")
print(f"el recorrido en aproximadamente {tiempo_predicho:.2f} minutos.")
print(f"\nEsto representa una reducci√≥n de {30.5 - tiempo_predicho:.2f} minutos respecto")
print(f"a un corredor que entrena solo 2 horas semanales (30.5 min).")
```
