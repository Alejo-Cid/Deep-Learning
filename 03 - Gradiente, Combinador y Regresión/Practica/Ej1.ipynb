Indique el vector gradiente para  = 1 e  = 1



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la funcin f(x,y) = x / (2x + 3y + 1)
def f(x, y):
    return x / (2*x**2 + 3*y**2 + 1)

# Derivada parcial respecto a x: f/x = (-2x + 3y + 1) / (2x + 3y + 1)
def df_dx(x, y):
    numerador = -2*x**2 + 3*y**2 + 1
    denominador = (2*x**2 + 3*y**2 + 1)**2
    return numerador / denominador

# Derivada parcial respecto a y: f/y = -6xy / (2x + 3y + 1)
def df_dy(x, y):
    numerador = -6*x*y
    denominador = (2*x**2 + 3*y**2 + 1)**2
    return numerador / denominador

# Punto dado
x0, y0 = 1, -1

# Calcular el gradiente en (1, -1)
gradiente = np.array([df_dx(x0, y0), df_dy(x0, y0)])

print(f"Punto: ({x0}, {y0})")
print(f"Valor de la funcin: f({x0}, {y0}) = {f(x0, y0):.4f}")
print(f"Vector gradiente f(1, -1) = ({gradiente[0]:.4f}, {gradiente[1]:.4f})")
print(f"Mdulo del gradiente: {np.linalg.norm(gradiente):.4f}")

# Crear la grfica
fig = plt.figure(figsize=(14, 5))

# Grfica 3D
ax1 = fig.add_subplot(121, projection='3d')

# Crear una malla de puntos
x_range = np.linspace(-2, 2, 50)
y_range = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x_range, y_range)
Z = f(X, Y)

# Superficie
ax1.plot_surface(X, Y, Z, alpha=0.6, cmap='viridis')

# Punto en la superficie
z0 = f(x0, y0)
ax1.scatter([x0], [y0], [z0], color='red', s=100, label=f'Punto ({x0}, {y0})')

# Vector gradiente en 3D (escalado para visualizacin)
scale = 0.5
ax1.quiver(x0, y0, z0, gradiente[0], gradiente[1], 0, 
           color='red', arrow_length_ratio=0.3, linewidth=2, label='Gradiente')

ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x,y)')
ax1.set_title('Superficie y Vector Gradiente')
ax1.legend()

# Grfica de contorno 2D
ax2 = fig.add_subplot(122)

# Curvas de nivel
contour = ax2.contour(X, Y, Z, levels=20, cmap='viridis')
ax2.clabel(contour, inline=True, fontsize=8)

# Punto
ax2.plot(x0, y0, 'ro', markersize=10, label=f'Punto ({x0}, {y0})')

# Vector gradiente
ax2.quiver(x0, y0, gradiente[0], gradiente[1], 
           color='red', scale=2, width=0.015, label='Gradiente')

ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Curvas de Nivel y Vector Gradiente')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.axis('equal')

plt.tight_layout()
plt.show()
```

### c)

```python
# EJERCICIO c)
print("\n" + "="*80)
print("EJERCICIO c) - Nueva posicin con alpha = 0.1")
print("="*80)

# Punto inicial
x_c, y_c = 0.20, -1
alpha = 0.1

# Calcular el gradiente en (0.20, -1)
grad_x_c = df_dx(x_c, y_c)
grad_y_c = df_dy(x_c, y_c)
gradiente_c = np.array([grad_x_c, grad_y_c])

print(f"\nPunto inicial: ({x_c}, {y_c})")
print(f"Valor inicial de la funcin: f({x_c}, {y_c}) = {f(x_c, y_c):.6f}")
print(f"\nVector gradiente: f({x_c}, {y_c}) = ({grad_x_c:.6f}, {grad_y_c:.6f})")
print(f"Alpha (tasa de aprendizaje): {alpha}")

# Actualizacin usando ASCENSO de gradiente (para maximizar)
# Nueva posicin = Posicin actual + alpha * gradiente
x_nuevo = x_c + alpha * grad_x_c
y_nuevo = y_c + alpha * grad_y_c

# Calcular el valor de la funcin en la nueva posicin
f_nuevo = f(x_nuevo, y_nuevo)

print("\n" + "-"*80)
print("NUEVA POSICIN:")
print("-"*80)
print(f"\nx_nuevo = {x_c} + {alpha} * {grad_x_c:.6f} = {x_nuevo:.6f}")
print(f"y_nuevo = {y_c} + {alpha} * {grad_y_c:.6f} = {y_nuevo:.6f}")
print(f"\nNueva posicin: ({x_nuevo:.6f}, {y_nuevo:.6f})")
print(f"Valor de la funcin en la nueva posicin: f({x_nuevo:.6f}, {y_nuevo:.6f}) = {f_nuevo:.6f}")

print("\n" + "-"*80)
print("COMPARACIN:")
print("-"*80)
print(f"Valor inicial:  {f(x_c, y_c):.6f}")
print(f"Valor nuevo:    {f_nuevo:.6f}")
print(f"Diferencia:     {f_nuevo - f(x_c, y_c):.6f}")
if f_nuevo > f(x_c, y_c):
    print(" La funcin CRECI (como se esperaba)")
else:
    print(" La funcin DECRECI")

# Visualizacin
fig = plt.figure(figsize=(14, 5))

# Grfica 3D
ax1 = fig.add_subplot(121, projection='3d')
x_range = np.linspace(-1, 1.5, 50)
y_range = np.linspace(-2, 0.5, 50)
X, Y = np.meshgrid(x_range, y_range)
Z = f(X, Y)

ax1.plot_surface(X, Y, Z, alpha=0.4, cmap='viridis')

# Punto inicial
z_inicial = f(x_c, y_c)
ax1.scatter([x_c], [y_c], [z_inicial], color='red', s=150, 
            label=f'Punto inicial ({x_c}, {y_c})', marker='o')

# Punto nuevo
ax1.scatter([x_nuevo], [y_nuevo], [f_nuevo], color='green', s=150, 
            label=f'Punto nuevo ({x_nuevo:.3f}, {y_nuevo:.3f})', marker='s')

# Vector de desplazamiento
ax1.quiver(x_c, y_c, z_inicial, 
           x_nuevo - x_c, y_nuevo - y_c, 0,
           color='blue', arrow_length_ratio=0.3, linewidth=3, 
           label='Desplazamiento')

ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x,y)')
ax1.set_title('Superficie y Desplazamiento')
ax1.legend()

# Grfica de contorno 2D
ax2 = fig.add_subplot(122)
contour = ax2.contour(X, Y, Z, levels=20, cmap='viridis')
ax2.clabel(contour, inline=True, fontsize=8)

# Punto inicial
ax2.plot(x_c, y_c, 'ro', markersize=12, label=f'Inicial ({x_c}, {y_c})')

# Punto nuevo
ax2.plot(x_nuevo, y_nuevo, 'gs', markersize=12, 
         label=f'Nuevo ({x_nuevo:.3f}, {y_nuevo:.3f})')

# Flecha de desplazamiento
ax2.arrow(x_c, y_c, x_nuevo - x_c, y_nuevo - y_c,
          head_width=0.05, head_length=0.05, fc='blue', ec='blue',
          label='Desplazamiento', linewidth=2)

# Vector gradiente (para referencia)
ax2.quiver(x_c, y_c, grad_x_c, grad_y_c,
           color='red', scale=3, width=0.008, alpha=0.5,
           label='f (direccin)')

ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Curvas de Nivel y Desplazamiento')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.axis('equal')

plt.tight_layout()
plt.show()
```

```python
# EJERCICIO f) - Encontrar el MNIMO usando descenso de gradiente
print("\n" + "="*80)
print("EJERCICIO f) - Bsqueda del MNIMO mediante DESCENSO de gradiente")
print("="*80)

# Hiperparmetros
alpha = 0.1           # Tasa de aprendizaje
max_iter = 1000       # Mximo de iteraciones
tolerancia = 1e-6     # Criterio de parada (cambio mnimo en la funcin)

# Punto inicial (puede ser aleatorio, elegimos uno cercano al origen)
x = 0.0
y = 0.0

# Historial para visualizacin
historial_x = [x]
historial_y = [y]
historial_f = [f(x, y)]

print(f"\nPunto inicial: ({x:.6f}, {y:.6f})")
print(f"Valor inicial: f({x:.6f}, {y:.6f}) = {f(x, y):.6f}")
print(f"\nParmetros:")
print(f"   Alpha (tasa de aprendizaje): {alpha}")
print(f"   Mximo de iteraciones: {max_iter}")
print(f"   Tolerancia: {tolerancia}")

# Algoritmo de descenso de gradiente
print("\n" + "-"*80)
print("EJECUTANDO DESCENSO DE GRADIENTE...")
print("-"*80)

for iter in range(max_iter):
    # Calcular gradiente en el punto actual
    grad_x = df_dx(x, y)
    grad_y = df_dy(x, y)
    
    # Actualizar posicin (DESCENSO: restamos el gradiente)
    x_nuevo = x - alpha * grad_x
    y_nuevo = y - alpha * grad_y
    
    # Calcular nuevo valor de la funcin
    f_nuevo = f(x_nuevo, y_nuevo)
    
    # Guardar en historial
    historial_x.append(x_nuevo)
    historial_y.append(y_nuevo)
    historial_f.append(f_nuevo)
    
    # Criterio de parada: si el cambio es muy pequeo
    cambio = abs(f_nuevo - f(x, y))
    if cambio < tolerancia:
        print(f"\nConvergencia alcanzada en iteracin {iter + 1}!")
        print(f"Cambio en f: {cambio:.10f} < {tolerancia}")
        break
    
    # Actualizar posicin
    x, y = x_nuevo, y_nuevo
    
    # Mostrar progreso cada 100 iteraciones
    if (iter + 1) % 100 == 0:
        print(f"Iteracin {iter + 1}: f({x:.6f}, {y:.6f}) = {f(x, y):.6f}")

# Resultados finales
print("\n" + "="*80)
print("RESULTADO FINAL:")
print("="*80)
print(f"\nMNIMO encontrado en:")
print(f"   Posicin: ({x:.6f}, {y:.6f})")
print(f"   Valor de la funcin: f({x:.6f}, {y:.6f}) = {f(x, y):.6f}")
print(f"\nIteraciones realizadas: {len(historial_f) - 1}")
print(f"Mdulo del gradiente final: {np.sqrt(grad_x**2 + grad_y**2):.10f}")

# Visualizacin del proceso
fig = plt.figure(figsize=(18, 5))

# Grfica 3D con trayectoria
ax1 = fig.add_subplot(131, projection='3d')
x_range = np.linspace(-2, 2, 50)
y_range = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x_range, y_range)
Z = f(X, Y)

ax1.plot_surface(X, Y, Z, alpha=0.3, cmap='viridis')

# Trayectoria del descenso
historial_z = [f(historial_x[i], historial_y[i]) for i in range(len(historial_x))]
ax1.plot(historial_x, historial_y, historial_z, 'r-', linewidth=2, label='Trayectoria')
ax1.scatter(historial_x[0], historial_y[0], historial_z[0], 
            color='green', s=150, marker='o', label='Inicio')
ax1.scatter(historial_x[-1], historial_y[-1], historial_z[-1], 
            color='red', s=150, marker='*', label='Mnimo')

ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x,y)')
ax1.set_title('Descenso de Gradiente - Vista 3D')
ax1.legend()

# Grfica de contorno con trayectoria
ax2 = fig.add_subplot(132)
contour = ax2.contour(X, Y, Z, levels=30, cmap='viridis')
ax2.clabel(contour, inline=True, fontsize=8)

# Trayectoria en 2D
ax2.plot(historial_x, historial_y, 'r-', linewidth=2, alpha=0.7)
ax2.plot(historial_x, historial_y, 'ro', markersize=4, alpha=0.5)
ax2.plot(historial_x[0], historial_y[0], 'go', markersize=12, label='Inicio')
ax2.plot(historial_x[-1], historial_y[-1], 'r*', markersize=20, label='Mnimo')

ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Trayectoria del Descenso')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.axis('equal')

# Grfica de convergencia
ax3 = fig.add_subplot(133)
ax3.plot(range(len(historial_f)), historial_f, 'b-', linewidth=2)
ax3.axhline(y=historial_f[-1], color='r', linestyle='--', 
            label=f'Mnimo = {historial_f[-1]:.6f}')
ax3.set_xlabel('Iteracin')
ax3.set_ylabel('f(x, y)')
ax3.set_title('Convergencia del Algoritmo')
ax3.grid(True, alpha=0.3)
ax3.legend()

plt.tight_layout()
plt.show()

# Informacin adicional
print("\n" + "-"*80)
print("INFORMACIN ADICIONAL:")
print("-"*80)
print(f"\nPrimeros 5 valores de f:")
for i in range(min(5, len(historial_f))):
    print(f"  Iteracin {i}: f = {historial_f[i]:.6f}")
print(f"\nltimos 5 valores de f:")
for i in range(max(0, len(historial_f)-5), len(historial_f)):
    print(f"  Iteracin {i}: f = {historial_f[i]:.6f}")
```
