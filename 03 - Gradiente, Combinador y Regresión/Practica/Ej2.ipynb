# Ejercicio 2
El precio de una vivienda está dado por dos factores, la proximidad a centros comerciales, financieros, vías
y demás que se mide a través del factor j, y de la condición misma de la vivienda que se mide por el factor k.

La relación entre estos factores y el precio de la vivienda en un barrio determinado está dada por:



donde el precio está dado en miles de dólares.

Utilice la técnica de descenso de gradiente para determinar los valores de j y k que minimizan el precio de
la vivienda.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

print("="*80)
print("EJERCICIO 2 - Minimización del precio de vivienda")
print("="*80)

# Definir la función de precio: P(j,k) = k² - 2jk + 2j² + 18j - 20k + 201
def P(j, k):
    """Precio de la vivienda en miles de dólares"""
    return k**2 - 2*j*k + 2*j**2 + 18*j - 20*k + 201

# Derivadas parciales
def dP_dj(j, k):
    """Derivada parcial de P respecto a j: ∂P/∂j = -2k + 4j + 18"""
    return -2*k + 4*j + 18

def dP_dk(j, k):
    """Derivada parcial de P respecto a k: ∂P/∂k = 2k - 2j - 20"""
    return 2*k - 2*j - 20

print("\nFUNCIÓN A MINIMIZAR:")
print("P(j,k) = k² - 2jk + 2j² + 18j - 20k + 201")
print("\nDERIVADAS PARCIALES:")
print("∂P/∂j = -2k + 4j + 18")
print("∂P/∂k = 2k - 2j - 20")

# Hiperparámetros del descenso de gradiente
alpha = 0.01          # Tasa de aprendizaje (pequeña para estabilidad)
max_iter = 10000      # Máximo de iteraciones
tolerancia = 1e-8     # Criterio de parada

# Punto inicial (puede ser aleatorio)
j = 0.0
k = 0.0

# Historial para visualización
historial_j = [j]
historial_k = [k]
historial_P = [P(j, k)]

print("\n" + "="*80)
print("CONFIGURACIÓN DEL ALGORITMO")
print("="*80)
print(f"Punto inicial: (j, k) = ({j:.4f}, {k:.4f})")
print(f"Precio inicial: P = ${P(j, k):.2f} mil")
print(f"Tasa de aprendizaje (alpha): {alpha}")
print(f"Máximo de iteraciones: {max_iter}")
print(f"Tolerancia: {tolerancia}")

# Algoritmo de descenso de gradiente
print("\n" + "="*80)
print("EJECUTANDO DESCENSO DE GRADIENTE...")
print("="*80)

for iter in range(max_iter):
    # Calcular gradiente
    grad_j = dP_dj(j, k)
    grad_k = dP_dk(j, k)
    
    # Actualizar posición (DESCENSO: restamos el gradiente)
    j_nuevo = j - alpha * grad_j
    k_nuevo = k - alpha * grad_k
    
    # Calcular nuevo precio
    P_nuevo = P(j_nuevo, k_nuevo)
    
    # Guardar historial
    historial_j.append(j_nuevo)
    historial_k.append(k_nuevo)
    historial_P.append(P_nuevo)
    
    # Criterio de parada
    cambio = abs(P_nuevo - P(j, k))
    if cambio < tolerancia:
        print(f"\n¡Convergencia alcanzada en iteración {iter + 1}!")
        print(f"Cambio en P: ${cambio:.10f} mil < ${tolerancia} mil")
        break
    
    # Actualizar valores
    j, k = j_nuevo, k_nuevo
    
    # Mostrar progreso
    if (iter + 1) % 1000 == 0:
        print(f"Iteración {iter + 1}: P({j:.4f}, {k:.4f}) = ${P(j, k):.4f} mil")

# Resultados finales
print("\n" + "="*80)
print("RESULTADO FINAL - PRECIO MÍNIMO")
print("="*80)
print(f"\nFactores óptimos:")
print(f"  • j (proximidad): {j:.6f}")
print(f"  • k (condición): {k:.6f}")
print(f"\nPrecio mínimo de la vivienda:")
print(f"  • P(j, k) = ${P(j, k):.4f} mil dólares")
print(f"  • P(j, k) = ${P(j, k)*1000:.2f} dólares")
print(f"\nIteraciones realizadas: {len(historial_P) - 1}")
print(f"Módulo del gradiente final: {np.sqrt(grad_j**2 + grad_k**2):.10f}")

# Verificación analítica (el mínimo se encuentra donde el gradiente es cero)
print("\n" + "="*80)
print("VERIFICACIÓN ANALÍTICA")
print("="*80)
print("\nPara encontrar el mínimo, igualamos el gradiente a cero:")
print("∂P/∂j = -2k + 4j + 18 = 0")
print("∂P/∂k = 2k - 2j - 20 = 0")
print("\nResolviendo el sistema:")
print("De la segunda ecuación: k = j + 10")
print("Sustituyendo en la primera: -2(j+10) + 4j + 18 = 0")
print("                           -2j - 20 + 4j + 18 = 0")
print("                            2j - 2 = 0")
print("                            j = 1")
print("                            k = 11")
j_analitico = 1.0
k_analitico = 11.0
P_analitico = P(j_analitico, k_analitico)
print(f"\nSolución analítica: j = {j_analitico}, k = {k_analitico}")
print(f"Precio analítico: P = ${P_analitico:.4f} mil")
print(f"\nDiferencia con solución numérica:")
print(f"  Δj = {abs(j - j_analitico):.10f}")
print(f"  Δk = {abs(k - k_analitico):.10f}")
print(f"  ΔP = ${abs(P(j,k) - P_analitico):.10f} mil")

# Visualización
fig = plt.figure(figsize=(18, 5))

# Gráfica 3D con trayectoria
ax1 = fig.add_subplot(131, projection='3d')
j_range = np.linspace(-5, 5, 50)
k_range = np.linspace(5, 15, 50)
J, K = np.meshgrid(j_range, k_range)
Z = P(J, K)

surf = ax1.plot_surface(J, K, Z, alpha=0.3, cmap='viridis')

# Trayectoria del descenso
historial_z = [P(historial_j[i], historial_k[i]) for i in range(len(historial_j))]
ax1.plot(historial_j, historial_k, historial_z, 'r-', linewidth=2, label='Trayectoria')
ax1.scatter(historial_j[0], historial_k[0], historial_z[0], 
            color='green', s=150, marker='o', label='Inicio', edgecolors='black')
ax1.scatter(historial_j[-1], historial_k[-1], historial_z[-1], 
            color='red', s=200, marker='*', label='Mínimo', edgecolors='black')

ax1.set_xlabel('j (proximidad)')
ax1.set_ylabel('k (condición)')
ax1.set_zlabel('P(j,k) [miles $]')
ax1.set_title('Superficie de Precio - Descenso de Gradiente')
ax1.legend()
fig.colorbar(surf, ax=ax1, shrink=0.5)

# Gráfica de contorno con trayectoria
ax2 = fig.add_subplot(132)
contour = ax2.contour(J, K, Z, levels=30, cmap='viridis')
ax2.clabel(contour, inline=True, fontsize=8)

# Trayectoria en 2D
ax2.plot(historial_j, historial_k, 'r-', linewidth=2, alpha=0.7, label='Trayectoria')
ax2.plot(historial_j, historial_k, 'ro', markersize=3, alpha=0.5)
ax2.plot(historial_j[0], historial_k[0], 'go', markersize=12, 
         label=f'Inicio ({historial_j[0]:.1f}, {historial_k[0]:.1f})', 
         markeredgecolor='black', markeredgewidth=1.5)
ax2.plot(historial_j[-1], historial_k[-1], 'r*', markersize=20, 
         label=f'Mínimo ({j:.2f}, {k:.2f})', 
         markeredgecolor='black', markeredgewidth=1.5)

ax2.set_xlabel('j (proximidad)')
ax2.set_ylabel('k (condición)')
ax2.set_title('Curvas de Nivel - Trayectoria del Descenso')
ax2.grid(True, alpha=0.3)
ax2.legend()

# Gráfica de convergencia
ax3 = fig.add_subplot(133)
ax3.plot(range(len(historial_P)), historial_P, 'b-', linewidth=2)
ax3.axhline(y=historial_P[-1], color='r', linestyle='--', 
            label=f'Mínimo = ${historial_P[-1]:.4f} mil')
ax3.set_xlabel('Iteración')
ax3.set_ylabel('Precio P(j,k) [miles $]')
ax3.set_title('Convergencia del Precio')
ax3.grid(True, alpha=0.3)
ax3.legend()

plt.tight_layout()
plt.show()

# Información adicional
print("\n" + "="*80)
print("EVOLUCIÓN DEL PRECIO")
print("="*80)
print(f"\nPrimeros 5 valores:")
for i in range(min(5, len(historial_P))):
    print(f"  Iteración {i}: P = ${historial_P[i]:.6f} mil")
print(f"\nÚltimos 5 valores:")
for i in range(max(0, len(historial_P)-5), len(historial_P)):
    print(f"  Iteración {i}: P = ${historial_P[i]:.6f} mil")

print("\n" + "="*80)
print("INTERPRETACIÓN DEL RESULTADO")
print("="*80)
print(f"\nPara minimizar el precio de la vivienda en este barrio:")
print(f"  • La proximidad (j) debe ser: {j:.4f}")
print(f"  • La condición (k) debe ser: {k:.4f}")
print(f"\nCon estos valores, el precio mínimo alcanzable es:")
print(f"  • ${P(j,k):.4f} mil dólares = ${P(j,k)*1000:.2f} dólares")
print(f"\nNota: Valores negativos de j podrían indicar alejamiento de centros")
print(f"      comerciales para reducir el precio.")
print("="*80)
```
